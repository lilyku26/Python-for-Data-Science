```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

定義函數 y = 20*x*sin(x) + x^2 + x^3

```python
def Lossfunction(x):
    L = 20*x*np.sin(x) + x**2 + x**3
    return L
def diffLoss(x):
    diffL = 20*x*np.cos(x) + 2*x + 3*(x**2)
    return diffL
```

定義梯度下降函數

```python
def gradient(x_ini, learning_rate, iteration):
    xwalk = np.zeros(iteration + 1)
    xwalk[0] = x_ini
    for i in range (iteration):
        x_new = xwalk[i] - (learning_rate)*(diffLoss(xwalk[i]))
        xwalk [i+1] = x_new
    return xwalk
```

初始值從9開始，進行迭代30000次，黑線為與真正最小值的誤差

```python
x = gradient(9, 0.05, 30000)
t = np.linspace(-10,10,100)
plt.figure(figsize=(10,10))
plt.plot(t, Lossfunction(t))
plt.scatter(x, Lossfunction(x), c=x, cmap = 'rainbow')
plt.vlines(x,-1000,Lossfunction(x))
print('It just find local minimum, x= ',float(x[10000]), 'Lossfunction value =', float(Lossfunction(x[10000])))
```

這種梯度下降法的最大缺點，就是會在找到local minimum就停滯擺盪
