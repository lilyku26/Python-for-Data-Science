```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

$$定義函數y = 50x - 100x^2 - 3x^3 +x^4 $$

```python
def Lossfunction(x):
    L = 50*x - 100*(x**2) - 3*x**3 + x**4 
    return L
def diffLoss(x):
    diffL = 50 - 200*x - 9*(x**2) + 4*(x**3)
    return diffL
```

$$畫出  Lossfunction(x)$$

```python
x = np.linspace(-10,10,100)
plt.figure(figsize=(10,10))
plt.plot(x, Lossfunction(x))
```

$$定義梯度下降函數$$

```python
def gradient(x_ini, learning_rate, iteration):
    xwalk = np.zeros(iteration + 1)
    xwalk[0] = x_ini
    for i in range (iteration):
        x_new = xwalk[i] - (learning_rate)*(diffLoss(xwalk[i]))
        xwalk [i+1] = x_new
    return xwalk
```

$$初始值從9開始，進行迭代30000次，黑線為與真正最小值的誤差$$

```python
x = gradient(0, 0.0001, 30000)
t = np.linspace(-10,10,100)
plt.figure(figsize=(10,10))
plt.plot(t, Lossfunction(t))
plt.scatter(x, Lossfunction(x), c=x, cmap = 'rainbow')
plt.vlines(x,min(Lossfunction(t)),Lossfunction(x), color = 'g')
print('It just find local minimum, x= ',float(x[10000]), 'Lossfunction value =', float(Lossfunction(x[10000])))
```

$$這種梯度下降法的缺點，就是會混淆local minimum和真正的minimum$$
